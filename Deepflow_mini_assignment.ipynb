```python
text_cols = ['name', 'location', 'rest_type', 'cuisines', 'listed_in(type)', 'listed_in(city)']
for col in text_cols:
    df[col] = df[col].str.title().str.strip()
```

```python
def standardize_cuisine_order(cuisine_string):
    items = [item.strip().title() for item in cuisine_string.split(',')]
    return ', '.join(sorted(items))

df['cuisines'] = df['cuisines'].apply(standardize_cuisine_order)
```

```python
df.head()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
display(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['Rating', 'approx_cost(for two people)']] = scaler.fit_transform(df[['Rating', 'approx_cost(for two people)']])
display(df.head())
```

```python
display(df['location'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['location_encoded'] = label_encoder.fit_transform(df['location'])

display(df.head())
```

```python
# Get the mapping of encoded labels to original locations
location_mapping = dict(zip(label_encoder.transform(label_encoder.classes_), label_encoder.classes_))
sorted_location_mapping = {k: location_mapping[k] for k in sorted(location_mapping)}

print("Mapping of Encoded Integer to Location:")
for encoded_value, location_name in sorted_location_mapping.items():
    print(f"{encoded_value}: {location_name}")
```

```python
location_counts = df['location'].value_counts()
display(location_counts)
```

## Grouping Dataset in Cuisine

---



```python
cuisine_summary = df.groupby('cuisines').agg(
    Average_Rating=('Rating', 'mean'),
    Total_Votes=('votes', 'sum'),
    Min_Cost=('approx_cost(for two people)', 'min'),
    Max_Cost=('approx_cost(for two people)', 'max')
)
display(cuisine_summary.head())
```

## Identify top cuisines

Based on the aggregated metrics, identify the top cuisines based on criteria like average rating and popularity (votes).


```python
top_cuisines_by_rating = cuisine_summary.sort_values(by='Average_Rating', ascending=False).head(10)
print("Top 10 Cuisines by Average Rating:")
display(top_cuisines_by_rating)

top_cuisines_by_votes = cuisine_summary.sort_values(by='Total_Votes', ascending=False).head(10)
print("\nTop 10 Cuisines by Popularity (Votes):")
display(top_cuisines_by_votes)
```

### Highlight best rated restaurants with their cuisine types

```python
top_rated_restaurants = df.sort_values(by='Rating', ascending=False).head(10)
display(top_rated_restaurants[['name', 'Rating', 'cuisines']])
```

```python
print("Analysis of Top Cuisines and User Preferences:")
print("\nCuisines appearing in both Top 10 by Rating and Top 10 by Votes:")
common_cuisines = set(top_cuisines_by_rating.index).intersection(set(top_cuisines_by_votes.index))
if common_cuisines:
    for cuisine in common_cuisines:
        print(f"- {cuisine}")
else:
    print("No common cuisines in the top 10 by both rating and votes.")

print("\nCuisines unique to Top 10 by Rating:")
unique_to_rating = set(top_cuisines_by_rating.index) - set(top_cuisines_by_votes.index)
for cuisine in unique_to_rating:
    print(f"- {cuisine}")

print("\nCuisines unique to Top 10 by Votes:")
unique_to_votes = set(top_cuisines_by_votes.index) - set(top_cuisines_by_rating.index)
for cuisine in unique_to_votes:
    print(f"- {cuisine}")

# 2. Examine cuisine_summary for patterns
print("\nPatterns in Cuisine Summary (based on top cuisines):")
print("\nTop 10 Cuisines by Average Rating Summary:")
display(cuisine_summary.loc[top_cuisines_by_rating.index])

print("\nTop 10 Cuisines by Popularity (Votes) Summary:")
display(cuisine_summary.loc[top_cuisines_by_votes.index])

# 3. Refer back to top_rated_restaurants
print("\nCuisines of Top 10 Highest-Rated Individual Restaurants:")
display(top_rated_restaurants[['name', 'Rating', 'cuisines']])
```

## Visualize key findings


```python
# 1. Bar plot for top 10 cuisines by average rating
plt.figure(figsize=(12, 6))
plt.bar(top_cuisines_by_rating.index, top_cuisines_by_rating['Average_Rating'], color='skyblue')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 2. Bar plot for top 10 cuisines by popularity (votes)
plt.figure(figsize=(12, 6))
plt.bar(top_cuisines_by_votes.index, top_cuisines_by_votes['Total_Votes'], color='lightcoral')
plt.title('Top 10 Cuisines by Popularity (Votes)')
plt.xlabel('Cuisine')
plt.ylabel('Total Votes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 3. Scatter plot showing the relationship between votes and Rating
plt.figure(figsize=(10, 6))
plt.scatter(df['votes'], df['Rating'], alpha=0.5, color='green')
plt.title('Relationship between Votes and Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()

# Display all plots
plt.show()
```
